````python
def PlotViolin(subset1, subset2, feature):
    fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
    ax1.violinplot(subset1[feature])
    ax1.set_title('{set} {feature}'.format(set=subset1['sentiments'][0], feature=feature))

    ax2.violinplot(subset2[feature])
    ax2.set_title('{set} {feature}'.format(set=subset2['sentiments'][0], feature=feature))
    plt.show()
````

````python
import csv
from sklearn.utils import Bunch

# edit 
def load_my_fancy_dataset_testing(my_data: pd.DataFrame):
      feature_names = my_data.columns
      feature_names = feature_names[:-1]
      X = []
      for feature in feature_names:
        feature_array = np.array(my_data[feature])
        X.append(feature_array)
      Y = my_data['sentiments']
      Y = Y.map({'angry': 0, 'soothing': 1})
      data = np.array(X)
      target = np.array(Y)
      return Bunch(data=data, target=target, feature_names=feature_names)
````

````python
trial = trial.drop(columns = ['id', 'title', 'artist', 'key', 'loudness', 'mode', 'instrumentalness', 'liveness', 'duration_ms', 'time_signature'])
````

````python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
````

````python
dataset = load_my_fancy_dataset_testing(trial)
X = dataset.data
Y = dataset.target
X = X.T
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
````

````python
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)
Y_predict = knn_model.predict(X_test)
````

````python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test, Y_predict)
plt.show()
````

````python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, Y_predict)
````

````python
from sklearn.metrics import precision_score
precision_score(y_test, Y_predict)
````

````python
# hyperparameter tuning

error_rate = []

for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    
    error_rate.append(np.mean(pred_i != y_test))
    
plt.figure(figsize=(10, 6))

plt.plot(range(1, 40), error_rate, color='blue', linestyle='--',
         markersize=10, markerfacecolor='red', marker='o')
plt.title('K versus Error rate')
plt.xlabel('K')
plt.ylabel('Error rate')
````
